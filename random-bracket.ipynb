```python
import random
```

```python
# We will establish the beginning bracket as a list #
bracket_list = [['UConn','Stetson'],['FAU','Northwestern'],['San Diego St','UAB'],['Auburn','Yale'],['BYU','Duquesne'],['Illinois','Morehead St'],['Washington St','Drake'],['Iowa Atate','S Dakota St'],['North Carolina','Wagner'],['Mississippi St','Michigan St'],['Saint Mary\'s','Grand Canyon'],['Alabama','Charleston'],['Clemson','New Mexico'],['Baylor','Colgate'],['Dayton','Nevada'],['Arizona','Long Beach St'],['Houston','Longwood'],['Nebraska','Texas A&M'],['Wisconsin','James Madison'],['Duke','Vermont'],['Texas Tech','NC State'],['Kentucky','Oakland'],['Forida','BOID/COL'],['Marquette','Western KY'],['Purdue','MTST'],['Utah State','TCU'],['Gonzaga','McNeese'],['Kansas','Sanford'],['South Carolina','Oregon'],['Creighton','Akron'],['Texas','UVA/CSU'],['Tennesse','Saint Peter\'s']]

round1_winners = []
for pair in bracket_list:
    winner = random.choice(pair)
    round1_winners.append(winner)

print("Tournament Brackets: \n" + str(round1_winners))

# Randomly pick each round and merge pairs into lists until only one team remains #
round2_bracket = []
i = 0
while i in range(len(round1_winners) - 1):
    round2_bracket.append([round1_winners[i], round1_winners[i + 1]])
    i += 2
    
print("\nRound 2 Bracket: \n" + str(round2_bracket))


round2_winners = []
for pair in round2_bracket:
    winner = random.choice(pair)
    round2_winners.append(winner)

print("\nRound 2 Winners: \n" + str(round2_winners))


round3_bracket = []
i = 0
while i in range(len(round2_winners) - 1):
    round3_bracket.append([round2_winners[i], round2_winners[i + 1]])
    i += 2
    
print("\nRound 3 Bracket: \n" + str(round3_bracket))


round3_winners = []
for pair in round3_bracket:
    winner = random.choice(pair)
    round3_winners.append(winner)

print("\nRound 3 Winners: \n" + str(round3_winners))



round4_bracket = []
i = 0
while i in range(len(round3_winners) - 1):
    round4_bracket.append([round3_winners[i], round3_winners[i + 1]])
    i += 2
    
print("\nRound 4 Bracket: \n" + str(round4_bracket))



round4_winners = []
for pair in round4_bracket:
    winner = random.choice(pair)
    round4_winners.append(winner)

print("\nRound 4 Winners: \n" + str(round4_winners))



round5_bracket = []
i = 0
while i in range(len(round4_winners) - 1):
    round5_bracket.append([round4_winners[i], round4_winners[i + 1]])
    i += 2
    
print("\nRound 5 Bracket: \n" + str(round5_bracket))



round5_winners = []
for pair in round5_bracket:
    winner = random.choice(pair)
    round5_winners.append(winner)

print("\nRound 5 Bracket: \n" + str(round5_winners))



winner = random.choice(round5_winners)
print("\nWinner: " + winner)
```
