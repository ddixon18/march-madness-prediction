```python
# Function to simulate a single game
def predict(team1, team2):
    result = 0
    mm2024_individual_teams_temp = mm2024_individual_teams.copy()
    team1_data = mm2024_individual_teams_temp[mm2024_individual_teams_temp['Team'] == name_to_id[team1]]
    team2_data = mm2024_individual_teams_temp[mm2024_individual_teams_temp['Team'] == name_to_id[team2]]
    input_data = team1_data.merge(team2_data, on = "YEAR", suffixes=['_Team1', '_Team2']).copy()    
    input_data.drop("YEAR", axis = 1)
    
    # Ensure the columns match out training data
    new_order = ['Team_Team1','Team_Team2', 'CONF_Team1', 'ADJOE_Team1', 'ADJDE_Team1',
       'BARTHAG_Team1', 'EFG_O_Team1', 'EFG_D_Team1', 'TOR_Team1', 'TORD_Team1', 'ORB_Team1', 'DRB_Team1', 'FTR_Team1', 'FTRD_Team1',
       '2P_O_Team1', '2P_D_Team1', '3P_O_Team1', '3P_D_Team1', 'ADJ_T_Team1', 'WAB_Team1', 'WIN_PCT_Team1', 'CONF_Team2', 'ADJOE_Team2', 'ADJDE_Team2',
       'BARTHAG_Team2', 'EFG_O_Team2', 'EFG_D_Team2', 'TOR_Team2',
       'TORD_Team2', 'ORB_Team2', 'DRB_Team2', 'FTR_Team2', 'FTRD_Team2',
       '2P_O_Team2', '2P_D_Team2', '3P_O_Team2', '3P_D_Team2', 'ADJ_T_Team2',
       'WAB_Team2', 'WIN_PCT_Team2']
    input_data = input_data[new_order]
    try:  
        result = rf.predict(np.array(input_data).reshape(1,-1))
    except ValueError:
        print(f'Either {team1} or {team2} not in list')
    if result == 1:
        winner = team1
    else:
        winner = team2
    print(f"Between {team1} and {team2} | {winner} won!")
    return winner
```

```python
def create_matchups(seeds):
  """
  Creates a dictionary mapping games to team matchups based on seed order.

  Args:
      seeds: Dictionary mapping team names to their seed.

  Returns:
      A dictionary mapping game numbers to team matchups.
  """
  sorted_seeds = dict(sorted(seeds.items(), key=lambda item: item[1]))
  teams = list(sorted_seeds.keys())
  num_games = len(teams) // 2
  matchups = {}
  for i in range(num_games):
    matchups[i+1] = (teams[i], teams[-i-1])  # Use integer index for game number
  return matchups
```

```python
# function to simulate tournament outcomes for a single region
def simulate_regionals(region):
    print('------------')
    matches = create_matchups(region)
    winners = {}
    for i, teams in matches.items():
        winner = predict(teams[0], teams[1])
        winners[winner] = i
        
    matches2 = create_matchups(winners)
    winners2 = {}
    for i, teams in matches2.items():
        winner = predict(teams[0], teams[1])
        winners2[winner] = i

    matches3 = create_matchups(winners2)
    winners3 = {}
    for i, teams in matches3.items():
        winner = predict(teams[0], teams[1])
        winners3[winner] = i

    matches4 = create_matchups(winners3)
    regional_winner = None
    for i, teams in matches4.items():
        regional_winner = predict(teams[0], teams[1])
    return regional_winner
```

```python
# function to predict the whole tournament

def simulate_tournament(regions):

    final_four = []
    
    for region in regions:
        final_four.append(simulate_regionals(region))


    print('-----SEMI FINALS-------')
    # semifinals
    w1 = predict(final_four[0], final_four[1])
    w2 = predict(final_four[2], final_four[3])

    #final
    print('-----FINALS-------')
    predict(w1, w2)
```

```python
# function to read in data from a file which contains our 2024 tournament teams, ordered by seed and separated by region
def read_teams_from_file(filename):
  regions = []
  current_region = []
  with open(filename, "r") as file:
    for line in file:
      line = line.strip()  # Remove leading/trailing whitespace
      if line:  # Skip empty lines
        current_region.append(line)
      elif current_region:  # If a region is complete, process it
        regions.append({team: i+1 for i, team in enumerate(current_region)})
        current_region = []
  if current_region:  # Process the last region
    regions.append({team: i+1 for i, team in enumerate(current_region)})
  return regions

# Get filename from user and read regions
filename = 'Data/24list.txt'
regions = read_teams_from_file(filename)

simulate_tournament(regions)
# print(bracket_results)
```
