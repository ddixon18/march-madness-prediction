```python
import pandas as pd
import numpy as np
import random
```

```python
cbb_table = pd.read_csv("Data/cbb.csv")
cbb_table.drop(['POSTSEASON'], axis=1, inplace=True)
cbb_table['WIN_PCT'] = cbb_table['W']/cbb_table['G']
cbb_table.drop(['G', 'W'], axis = 1, inplace = True)

# Check for names ending with '.'
mask = cbb_table['TEAM'].str.endswith('.')

# Remove the trailing '.' using 'str.rstrip'
cbb_table.loc[mask, 'TEAM'] = cbb_table.loc[mask, 'TEAM'].str.rstrip('.')
cbb_table = cbb_table.rename(columns={'TEAM': 'Team1'})
cbb_table = cbb_table.rename(columns={'YEAR': 'Year'})

cbb_table.drop(["SEED"], axis = 1, inplace = True)



cbb_table.head()
```

```python
tourn_table = pd.read_csv("Data/MNCAATourneyCompactResults.csv")
tourn_table.head()
```

```python
teams_table = pd.read_csv("Data/MTeams.csv")
teams_table

# Create easy mapping from integer TeamID to string Team Name
id_to_name = {}
name_to_id = {}
for idx, row in teams_table.iterrows():
    id_to_name[row["TeamID"]] = row["TeamName"]
    name_to_id[row["TeamName"]] = row["TeamID"]
    
# creating Team 1 and Team 2 columns
tourn_table['Team1'] = ""
tourn_table['Team2'] = ""
tourn_table['WTeamName'] = ""

for idx, row in tourn_table.iterrows():
    TEAMS = [row["WTeamID"], row["LTeamID"]]
    team1 = random.choice(TEAMS)
    team2 = TEAMS[0] if TEAMS[0] != team1 else TEAMS[1]
    tourn_table.at[idx, "Team1"] = id_to_name[team1]
    tourn_table.at[idx, "Team2"] = id_to_name[team2]
    tourn_table.at[idx, "WTeamName"] = id_to_name[row["WTeamID"]]
    
for idx, row in tourn_table.iterrows():
    TEAMS = [row["WTeamID"], row["LTeamID"]]
    team1 = random.choice(TEAMS)
    team2 = TEAMS[0] if TEAMS[0] != team1 else TEAMS[1]
    tourn_table.at[idx, "Team1"] = id_to_name[team1]
    tourn_table.at[idx, "Team2"] = id_to_name[team2]
    tourn_table.at[idx, "WTeamName"] = id_to_name[row["WTeamID"]]

tourn_table.drop(["LTeamID"], axis = 1, inplace = True)
tourn_table.drop(["NumOT"], axis = 1, inplace = True)
tourn_table = tourn_table.rename(columns={'Season': 'Year'})


tourn_table
```

```python
## Machine Learning -- yay :) ##

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics

labels = np.array(merged_stats_comp['WTeamName'])
le = LabelEncoder()
mm2024_merged_comp['CONF_Team1'] = le.fit_transform(mm2024_merged_comp['CONF_Team1'])
mm2024_merged_comp['CONF_Team2'] = le.fit_transform(mm2024_merged_comp['CONF_Team2'])



mm2024_merged_comp
```

```python
for idx, row in merged_stats_comp.iterrows():
    merged_stats_comp.at[idx, "Team1"] = name_to_id[row["Team1"]]
    merged_stats_comp.at[idx, "Team2"] = name_to_id[row["Team2"]]
    merged_stats_comp.at[idx, "WTeamName"] = name_to_id[row["WTeamName"]]
    
for idx, row in mm2024_merged_comp.iterrows():
    mm2024_merged_comp.at[idx, "Team1"] = name_to_id[row["Team1"]]
    mm2024_merged_comp.at[idx, "Team2"] = name_to_id[row["Team2"]]

merged_stats_comp
```

```python
train_features, test_features, train_labels, test_labels = train_test_split(merged_stats_comp, labels, test_size = 0.2, random_state = 42)

# Instantiate model with 5000 decision trees
rf = RandomForestClassifier(n_estimators = 5000, random_state = 42)
# Train the model on training data
rf.fit(train_features.values, train_labels)
```

```python
predictions = rf.predict(test_features)

# Checking Accuracy
print("Accuracy:",metrics.accuracy_score(test_labels, predictions))
```
